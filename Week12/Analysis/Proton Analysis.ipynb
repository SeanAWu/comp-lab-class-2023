```python
import os
import matplotlib.pyplot as plt

# Function to read data from the file
def read_data(file_path, file_name):
    step, time, kin, temp, pot, cons_qty, used_time = [], [], [], [], [], [], []

    file_pathname = os.path.join(file_path, file_name)

    with open(file_pathname, 'r') as file:
        next(file)  # Skip the header line
        for line in file:
            data = line.split()
            step.append(float(data[0]))
            time.append(float(data[1]))
            kin.append(float(data[2]))
            temp.append(float(data[3]))
            pot.append(float(data[4]))
            cons_qty.append(float(data[5]))
            used_time.append(float(data[6]))

    return step, time, kin, temp, pot, cons_qty, used_time

# Function to plot energy vs. time on separate figures
def plot_energy_vs_time(time, energy, label, ylabel, title):
    plt.figure(figsize=(10, 6))
    plt.plot(time, energy, label=label)
    plt.xlabel('Time [fs]')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Main script
file_path = '../Runs-Proton/'
file_name = 'proton_aimd_1ns_every0.5fs-1.ener'
step, time, kin, temp, pot, cons_qty, used_time = read_data(file_path, file_name)

# Plot Kinetic Energy
plot_energy_vs_time(time, kin, 'Kinetic Energy', 'Energy [a.u.]', 'Kinetic Energy vs. Time')

# Plot Potential Energy
plot_energy_vs_time(time, pot, 'Potential Energy', 'Energy [a.u.]', 'Potential Energy vs. Time')

# Plot Conserved Quantity
plot_energy_vs_time(time, cons_qty, 'Conserved Energy', 'Energy [a.u.]', 'Conserved Energy vs. Time')
```

```python
import MDAnalysis as md
import numpy as np
import matplotlib.pyplot as plt
from MDAnalysis.analysis.rdf import InterRDF

# Create a Universe object without a topology file
u = md.Universe('../Runs-Proton/proton_aimd_1ns_every0.5fs-pos-1.xyz')

box_vectors = [12.42, 12.42, 12.42, 90, 90, 90]
u.dimensions = box_vectors

# Select oxygen atoms
oxygen_atoms = u.select_atoms('name O')

# Set up the interRDF calculation
rdf_calculation = InterRDF(oxygen_atoms, oxygen_atoms, nbins=100, range=(0.0, 10.0))

# Set the maximum number of frames
max_frames = 1

# Loop through the trajectory frames and calculate RDF for the first 100 frames
for ts in u.trajectory[:max_frames]:
    rdf_calculation.run()

# Get the RDF data
rdf_data = rdf_calculation.rdf

# Scale the RDF data by the number of frames and the number of oxygen atoms
num_frames = len(u.trajectory[:max_frames])
num_oxygen_atoms = len(oxygen_atoms)
rdf_data_scaled = rdf_data / (num_frames * num_oxygen_atoms)

# Get the bin edges and calculate the bin centers
bin_edges = rdf_calculation.bins
bin_centers = (bin_edges[1:] + bin_edges[:-1]) / 2

# Ensure that rdf_data_scaled and bin_centers have the same dimensions
rdf_data_scaled = rdf_data_scaled[:len(bin_centers)]

# Plot the scaled RDF
plt.plot(bin_centers, rdf_data_scaled)
plt.xlabel('Distance (nm)')
plt.ylabel('RDF')
plt.ylim(0,0.05)
plt.title('Radial Distribution Function (RDF) of Oxygen Atoms')
plt.show()
```
