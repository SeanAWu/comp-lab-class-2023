```python
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt

def calculate_rmsd(pdb_file1, pdb_file2, atom_selection):
    traj1 = md.load(pdb_file1)
    traj2 = md.load(pdb_file2)

    atom_indices = None
    if atom_selection == "alpha-carbons":
        atom_indices = traj1.top.select("name CA")
    elif atom_selection == "backbone":
        atom_indices = traj1.top.select("backbone")
    elif atom_selection == "side-chains":
        atom_indices = traj1.top.select("not backbone and not water")

    rmsd_values = md.rmsd(traj1, traj2, atom_indices=atom_indices)

    return rmsd_values

def plot_rmsd_bar_chart(pdb_file1, pdb_file2):
    atom_selections = ["alpha-carbons", "backbone", "side-chains"]
    rmsd_data = {selection: calculate_rmsd(pdb_file1, pdb_file2, selection) for selection in atom_selections}

    plt.figure(figsize=(8, 6))
    plt.bar(atom_selections, [np.mean(rmsd_data[selection]) for selection in atom_selections], yerr=[np.std(rmsd_data[selection]) for selection in atom_selections])
    plt.xlabel("Atom Selection")
    plt.ylabel("RMSD")
    plt.title("Difference between Greene and PDB (UBQ)")
    plt.show()

if __name__ == "__main__":
    pdb_file1 = "UBQgreene.pdb"
    pdb_file2 = "1UBQ.pdb"
    plot_rmsd_bar_chart(pdb_file1, pdb_file2)
```
