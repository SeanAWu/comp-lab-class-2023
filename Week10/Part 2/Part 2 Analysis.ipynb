```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Lennard-Jones potential function
def lj(r, sigma, eps):
    return 4 * eps * ((sigma / r)**12 - (sigma / r)**6)

# Provided data points
data = np.array([
    (3.5, 0.168852),
    (4.0, -0.123122),
    (4.5, -0.095162),
    (5.0, -0.055927),
    (5.5, -0.032054),
    (6.0, -0.018843)
])

# Extracting data into separate arrays
r_data, E_int_data = data[:, 0], data[:, 1]

# Use curve_fit with modified initial parameter values
initial_params = (3.8, 0.1)  # Adjust these values based on your understanding of the data
optimal_params, covariance = curve_fit(lj, r_data, E_int_data, p0=initial_params)

# Print the optimal parameters
sigma_optimal, eps_optimal = optimal_params
print("Optimal Sigma:", sigma_optimal)
print("Optimal Epsilon:", eps_optimal)

# Generate a curve using the optimal parameters for visualization
r_fit = np.linspace(min(r_data), max(r_data), 100)
E_int_fit = lj(r_fit, sigma_optimal, eps_optimal)

# Plotting the data and the fitted curve
plt.scatter(r_data, E_int_data, label="Data")
plt.plot(r_fit, E_int_fit, label="Fitted Curve", color='red', linestyle = 'dashed')
plt.xlabel('r (Angstrom)')
plt.ylabel('E_int (kcal/mol)')
plt.legend()
plt.show()
```
