```python
import matplotlib.pyplot as plt
import numpy as np

# Define the file path for temperature data
temp_file_path = '/scratch/work/courses/CHEM-GA-2671-2023fa/students/sw4384/comp-lab-class-2023/Week3/Input/temperature.xvg'

# Initialize empty lists to store temperature data
time_data = []
temperature_data = []

# Read data from the temperature XVG file
with open(temp_file_path, 'r') as temp_file:
    temp_lines = temp_file.readlines()

# Extract the relevant data columns
for temp_line in temp_lines:
    if temp_line.startswith(('#', '@')):  # Skip comment lines and headers
        continue
    temp_columns = temp_line.strip().split()
    if len(temp_columns) >= 2:
        time_data.append(float(temp_columns[0]))
        temperature_data.append(float(temp_columns[1]))

# Create the temperature vs. time plot
plt.figure(figsize=(8, 6))
plt.plot(time_data, temperature_data, label='Temperature', linewidth=2)
plt.xlabel('Time (ps)')  # X-axis label
plt.ylabel('Temperature (K)')  # Y-axis label
plt.title('Temperature vs Time')  # Plot title
plt.legend()
plt.grid(True)

# Show the temperature plot
plt.show()
```
