```python
import matplotlib.pyplot as plt
import numpy as np

# Define the file path
file_path = '/scratch/work/courses/CHEM-GA-2671-2023fa/students/sw4384/comp-lab-class-2023/Week3/Input/potential.xvg'

# Initialize empty lists to store data
x_data = []
y_data = []

# Read data from the XVG file
with open(file_path, 'r') as f:
    lines = f.readlines()

# Extract the relevant data columns
for line in lines:
    if line.startswith(('#', '@')):  # Skip comment lines and headers
        continue
    columns = line.strip().split()
    if len(columns) >= 2:
        x_data.append(float(columns[0]))
        y_data.append(float(columns[1]))

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x_data, y_data, label='Potential', linewidth=2)
plt.xlabel('Energy Minimization Step')  # X-axis label
plt.ylabel('Potential Energy (kJ/mol)')  # Y-axis label
plt.title('Energy Minimization')  # Plot title
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
