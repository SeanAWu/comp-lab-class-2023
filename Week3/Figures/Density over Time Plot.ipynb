```python
import matplotlib.pyplot as plt
import numpy as np

# Define the file path for density data
density_file_path = '/scratch/work/courses/CHEM-GA-2671-2023fa/students/sw4384/comp-lab-class-2023/Week3/Input/density.xvg'

# Initialize empty lists to store density data
time_data_density = []
density_data = []

# Read data from the density XVG file
with open(density_file_path, 'r') as density_file:
    density_lines = density_file.readlines()

# Extract the relevant data columns
for density_line in density_lines:
    if density_line.startswith(('#', '@')):  # Skip comment lines and headers
        continue
    density_columns = density_line.strip().split()
    if len(density_columns) >= 2:
        time_data_density.append(float(density_columns[0]))
        density_data.append(float(density_columns[1]))

# Create the density vs. time plot
plt.figure(figsize=(8, 6))
plt.plot(time_data_density, density_data, label='Density', linewidth=2)
plt.xlabel('Time (ps)')  # X-axis label
plt.ylabel('Density (kg/m³)')  # Y-axis label with kg/m³ unit
plt.title('Density vs Time')  # Plot title
plt.legend()
plt.grid(True)

# Show the density plot
plt.show()
```
