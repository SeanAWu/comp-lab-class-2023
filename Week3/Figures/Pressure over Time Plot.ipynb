```python
import matplotlib.pyplot as plt
import numpy as np

# Define the file path for pressure data
pressure_file_path = '/scratch/work/courses/CHEM-GA-2671-2023fa/students/sw4384/comp-lab-class-2023/Week3/Input/pressure.xvg'

# Initialize empty lists to store pressure data
time_data_pressure = []
pressure_data = []

# Read data from the pressure XVG file
with open(pressure_file_path, 'r') as pressure_file:
    pressure_lines = pressure_file.readlines()

# Extract the relevant data columns
for pressure_line in pressure_lines:
    if pressure_line.startswith(('#', '@')):  # Skip comment lines and headers
        continue
    pressure_columns = pressure_line.strip().split()
    if len(pressure_columns) >= 2:
        time_data_pressure.append(float(pressure_columns[0]))
        pressure_data.append(float(pressure_columns[1]))

# Create the pressure vs. time plot
plt.figure(figsize=(8, 6))
plt.plot(time_data_pressure, pressure_data, label='Pressure', linewidth=2)
plt.xlabel('Time (ps)')  # X-axis label
plt.ylabel('Pressure (bar)')  # Y-axis label
plt.title('Pressure vs Time')  # Plot title
plt.legend()
plt.grid(True)

# Show the pressure plot
plt.show()
```
