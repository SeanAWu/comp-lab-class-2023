```python
import matplotlib.pyplot as plt
import numpy as np

# Define the file path for RMSD data
rmsd_file_path = '/scratch/work/courses/CHEM-GA-2671-2023fa/students/sw4384/comp-lab-class-2023/Week3/Analysis/rmsd_xtal.xvg'

# Initialize empty lists to store RMSD data
time_data_rmsd = []
rmsd_data = []

# Read data from the RMSD XVG file
with open(rmsd_file_path, 'r') as rmsd_file:
    rmsd_lines = rmsd_file.readlines()

# Extract the relevant data columns
for rmsd_line in rmsd_lines:
    if rmsd_line.startswith(('#', '@')):  # Skip comment lines and headers
        continue
    rmsd_columns = rmsd_line.strip().split()
    if len(rmsd_columns) >= 2:
        time_data_rmsd.append(float(rmsd_columns[0]))
        rmsd_data.append(float(rmsd_columns[1]))

# Create the RMSD vs. time plot
plt.figure(figsize=(8, 6))
plt.plot(time_data_rmsd, rmsd_data, label='RMSD', linewidth=2)
plt.xlabel('Time (ps)')  # X-axis label
plt.ylabel('RMSD (nm)')  # Y-axis label with nm unit
plt.title('RMSD')  # Plot title
plt.legend()
plt.grid(True)

# Show the RMSD plot
plt.show()

```
