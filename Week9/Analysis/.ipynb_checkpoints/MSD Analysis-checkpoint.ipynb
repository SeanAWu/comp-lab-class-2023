```python
import MDAnalysis as mda
import matplotlib.pyplot as plt
from MDAnalysis.analysis.rdf import InterRDF

# Load your universe (replace 'your_trajectory.xyz' and 'your_topology.pdb' with your actual file names)
universe = mda.Universe('../Data/equil/kalj_T0.5_n360_v300_10000000_1.lammpstrj', '../Data/prod/kalj_T0.5_n360_v300_prod_10000000_1.unwrap.dcd', topology_format="LAMMPSDUMP")

# Define selections
selection_type_1 = universe.select_atoms('type 1')
selection_type_2 = universe.select_atoms('type 2')

# Compute RDF for type 1 and type 1
rdf_1_1 = mda.analysis.rdf.InterRDF(selection_type_1, selection_type_1, nbins=100, range=(0, 10.0))
rdf_1_1.run()

# Compute RDF for type 2 and type 2
rdf_2_2 = mda.analysis.rdf.InterRDF(selection_type_2, selection_type_2, nbins=100, range=(0, 10.0))
rdf_2_2.run()

# Compute RDF for type 1 and type 2
rdf_1_2 = mda.analysis.rdf.InterRDF(selection_type_1, selection_type_2, nbins=100, range=(0, 10.0))
rdf_1_2.run()

# Plotting
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].plot(rdf_1_1.results.bins, rdf_1_1.results.rdf, label='Type 1 and Type 1')
axs[0].set_title('Type 1 and Type 1 RDF')

axs[1].plot(rdf_2_2.results.bins, rdf_2_2.results.rdf, label='Type 2 and Type 2')
axs[1].set_title('Type 2 and Type 2 RDF')

axs[2].plot(rdf_1_2.results.bins, rdf_1_2.results.rdf, label='Type 1 and Type 2')
axs[2].set_title('Type 1 and Type 2 RDF')

for ax in axs:
    ax.set_xlabel('Distance (nm)')
    ax.set_ylabel('RDF')
    ax.legend()

plt.tight_layout()
plt.show()
```
