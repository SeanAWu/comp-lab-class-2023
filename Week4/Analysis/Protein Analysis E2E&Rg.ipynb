```python
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt

# Load the trajectory
trajectory_file = '/scratch/work/courses/CHEM-GA-2671-2023fa/students/sw4384/comp-lab-class-2023/Week4/Analysis/wrappedprotein-Copy1.xtc'
topology_file = '/scratch/work/courses/CHEM-GA-2671-2023fa/students/sw4384/comp-lab-class-2023/Week4/Analysis/2jof-Copy1.pdb'

traj = md.load(trajectory_file, top=topology_file)

# Compute the distance between the first and last atom
def end_end(traj):
    first = traj.xyz[:, 0, :]
    last = traj.xyz[:, -1, :]
    dist = np.sqrt(np.sum((first - last) ** 2, axis=1))
    return dist

# Compute the distance between the first and last atom
dist = end_end(traj)

# Calculate the radius of gyration
Rg = md.compute_rg(traj)

# Create two separate subplots for distance vs. time and Rg vs. time
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(8, 6))

# Plot distance on the first subplot
ax1.plot(traj.time, dist, label='Distance', color='b')
ax1.set_ylabel('Distance', color='b')
ax1.tick_params('y', colors='b')

# Plot Rg on the second subplot
ax2.plot(traj.time, Rg, label='Rg', color='r')
ax2.set_xlabel('Time (ps)')
ax2.set_ylabel('Rg', color='r')
ax2.tick_params('y', colors='r')

# Add legends to each subplot
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Add titles to each subplot
ax1.set_title('Distance vs. Time')
ax2.set_title('Rg vs. Time')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
