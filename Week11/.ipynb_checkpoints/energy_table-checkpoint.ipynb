```python
import rdkit
from rdkit import Chem
from rdkit.Chem import AllChem as Chem

from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
```

```python
# Open the file in read mode
with open('smiles.csv', 'r') as file:
    # Skip the first line (header)
    next(file)

    # Initialize arrays to store names and SMILES
    names = []
    smiles = []

    # Iterate through each line in the file
    for line in file:
        # Split the line using a comma as the delimiter
        parts = line.strip().split(',')

        # Extract name and SMILES notation
        smile = parts[0]
        name = parts[1]

        # Append data to the arrays
        names.append(name)
        smiles.append(smile)

# Print the extracted data or use it as needed
for name, smile in zip(names, smiles):
    print(f"{name}: {smile}")
```

```python
for i in smiles:
    print(i)
```

```python
mols = [Chem.MolFromSmiles(smi) for smi in smiles]
Draw.MolsToGridImage(mols, molsPerRow=5, subImgSize=(200, 200))
```

```python
import os
import sys
import pandas as pd
import numpy as np
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole

sys.path.append('/Users/iwatobipen/develop/chemoenv/psikit/psikit/')

from psikit import Psikit
```

```python
# List of SMILES notations
smiles_list = smiles  # Replace with your SMILES notations
energy_list = []

# Iterate over the SMILES list
for smile in smiles_list:
    # Initialize Psikit for each molecule
    pk=Psikit(debug=True,threads=1,memory=12)

    # Set the molecular geometry using SMILES notation
    pk.read_from_smiles(smile)

    # Optimize the geometry
    pk.optimize()

    # Get the optimized energy
    energy = pk.energy()
    
    # Add energy to list
    energy_list.append(energy)

    # Print the result for each molecule
    print(f"SMILES: {smiles}, Optimized Energy: {energy} Hartree")
```
