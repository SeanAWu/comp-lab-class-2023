```python
import numpy as np
import matplotlib.pyplot as plt

def read_colvar_file(filename):
    data = np.loadtxt(filename)
    return data

def calculate_dihedrals(data):
    # Assuming you want to plot values from columns 1 and 2
    dihedral_columns = [1, 2]

    dihedrals = []
    for col in dihedral_columns:
        dihedral = data[:, col]
        dihedrals.append(dihedral)
    return dihedrals

def plot_ramachandran(dihedrals):
    plt.figure(figsize=(8, 6))
    plt.scatter(dihedrals[0], dihedrals[1], alpha=0.5, s=10)  # Scatter plot of dihedral angles

    plt.xlabel('Dihedral Angle 1 (degrees)')
    plt.ylabel('Dihedral Angle 2 (degrees)')
    plt.title('Ramachandran-like Plot')
    plt.grid(True)
    plt.show()

def print_standard_deviation(dihedrals):
    for i, dihedral in enumerate(dihedrals):
        std_dev = np.std(dihedral)
        print(f"Standard Deviation of Dihedral Angle {i + 1}: {std_dev:.4f} degrees")

if __name__ == '__main__':
    colvar_filename = 'COLVAR'
    data = read_colvar_file(colvar_filename)
    dihedrals = calculate_dihedrals(data)
    plot_ramachandran(dihedrals)
    print_standard_deviation(dihedrals)
```
