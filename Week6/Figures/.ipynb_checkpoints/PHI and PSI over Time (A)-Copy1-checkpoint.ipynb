```python
import matplotlib.pyplot as plt

# Read the COLVAR file and extract data
time = []
phi = []
psi = []

with open("COLVARPHI&PSI", "r") as f:
    for line in f:
        if line.startswith("#"):
            continue  # Skip comment lines
        parts = line.split()
        current_time = float(parts[0])  # Assuming time is in the first column

        if current_time > 2000:
            break  # Exit the loop if time exceeds 2000

        time.append(current_time)
        phi.append(float(parts[1]))  # Assuming phi is in the second column
        psi.append(float(parts[2]))  # Assuming psi is in the third column

# Create two separate scatterplots with smaller dots for data up to time=2000
plt.figure(1)
plt.subplot(211)
plt.scatter(time, phi, s=1)  # Smaller dots (change 's' value)
plt.xlabel("Time")
plt.ylabel("Phi")
plt.title("Phi Over Time")

plt.subplot(212)
plt.scatter(time, psi, s=1)  # Smaller dots (change 's' value)
plt.xlabel("Time")
plt.ylabel("Psi")
plt.title("Psi Over Time")

plt.tight_layout()  # To prevent overlapping labels and titles
plt.show()
```
