```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the text file
data = np.loadtxt('fes.dat')

# Extract columns from the data
phi = data[:, 0]
psi = data[:, 1]
free_energy = data[:, 2]

# Reshape the data to a 2D grid
n_phi = len(np.unique(phi))
n_psi = len(np.unique(psi))
phi = phi.reshape((n_phi, n_psi))
psi = psi.reshape((n_phi, n_psi))
free_energy = free_energy.reshape((n_phi, n_psi))

# Create a plot using imshow
plt.imshow(free_energy, cmap='viridis', origin='lower', extent=(phi.min(), phi.max(), psi.min(), psi.max()))
plt.colorbar(label='Free Energy')

# Add labels and title
plt.xlabel('Phi')
plt.ylabel('Psi')
plt.title('Free Energy Landscape')

# Show the plot
plt.show()
```
