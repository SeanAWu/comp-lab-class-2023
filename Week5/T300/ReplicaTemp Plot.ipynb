```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the "replica_temp.xvg" file
data = np.loadtxt('replica_temp.xvg', comments=['#', '@'])

# Assuming the first column contains time (or frame number)
time = data[:, 0]

# Assuming the second column contains the data you want to plot
data_to_plot = data[:, 1]

# Create a time series plot, plotting every 100 frames
plot_interval = 100
plt.plot(time[::plot_interval], data_to_plot[::plot_interval])
plt.xlabel('Time (or Frame Number)')
plt.ylabel('Data')
plt.title('Data vs. Frame Number Every 100 Frames')
plt.grid(True)
plt.show()
```
