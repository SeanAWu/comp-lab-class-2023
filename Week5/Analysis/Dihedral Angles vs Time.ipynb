```python
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt

# Load your MD trajectory and topology
traj = md.load('adpsim.trr', top='adpsim.gro')

# Define phi and psi atom indices for Alanine dipeptide within the range 0 to 22
phi_indices = [4, 6, 10, 12]
psi_indices = [6, 10, 12, 18]

# Calculate phi and psi dihedral angles
phi_angles = md.compute_dihedrals(traj, [phi_indices])
psi_angles = md.compute_dihedrals(traj, [psi_indices])

# Convert angles from radians to degrees
phi_degrees = np.rad2deg(phi_angles)
psi_degrees = np.rad2deg(psi_angles)

# Create a time array for x-axis
time = np.arange(len(phi_angles))

# Create two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

# Plot phi dihedral angle
ax1.plot(time, phi_degrees, label='Phi ()', linewidth=2)
ax1.set_ylabel('Dihedral Angle ()')
ax1.legend()
ax1.set_title('Phi Dihedral Angle for Alanine dipeptide')

# Plot psi dihedral angle
ax2.plot(time, psi_degrees, label='Psi ()', linewidth=2)
ax2.set_xlabel('Time Step')
ax2.set_ylabel('Dihedral Angle ()')
ax2.legend()
ax2.set_title('Psi Dihedral Angle for Alanine dipeptide')

# Add a common title
plt.suptitle('Phi and Psi Dihedral Angles for Alanine dipeptide')
plt.grid(True)
plt.show()
```
